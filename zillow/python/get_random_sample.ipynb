```python
import pandas as pd
import numpy as np

# Combine properties and sales data
# Some properties were sold in both 2016 and 2017 so mapping datasets within in a year preserves relationships
# 2016
properties_2016 = pd.read_table('/Users/David/Downloads/properties_2016.csv', sep = ',')
train_2016 = pd.read_table('/Users/David/Downloads/train_2016_v2.csv', sep = ',')
training_2016 = train_2016.merge(properties_2016, on='parcelid', how='left')
training_2016.shape
```

```python
# 2017
properties_2017 = pd.read_table('/Users/David/Downloads/properties_2017.csv', sep = ',')
train_2017 = pd.read_table('/Users/David/Downloads/train_2017.csv', sep = ',')
training_2017 = train_2017.merge(properties_2017, on='parcelid', how='left')
training_2017.shape
```

```python
# Combine the years into a complete set
frames = [training_2016, training_2017]
training_all = pd.concat(frames, ignore_index=True)
training_all.shape
```

```python
# Saving superset for future reference
training_all.to_csv("/Users/David/Desktop/Zillow/training_all.tsv", sep='\t', index=False)

# Add a random value column to the superset for sampling
total_rows = training_all.shape[0]
training_all['random_value'] = pd.Series(np.random.randn(total_rows), index=training_all.index)
randomized_training_all = training_all.sort_values(by=['random_value'])
# Taking about 10% of superset for now
training_sample = randomized_training_all.head(17000)
training_sample.to_csv("/Users/David/Desktop/Zillow/training_sample.tsv", sep='\t', index=False)
```
